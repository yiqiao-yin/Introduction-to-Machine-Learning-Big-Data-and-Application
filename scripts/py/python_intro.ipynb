### Dropping Columns

```python
# method 1
cols = [c for c in df.columns if '7day_avg' in c.lower() or 'probable' in c.lower()]
cols
```

```python
df.drop(cols, axis=1)
```

```python
# method 2
df = df.loc[:,~df.columns.str.contains('7day_avg', case=False)]
df = df.loc[:,~df.columns.str.contains('probable', case=False)]
df
```

### Adding columns

```python
df['TOTAL_CASES'] = df['CASE_COUNT'].cumsum()
df['TOTAL_HOSPITALIZED'] = df['HOSPITALIZED_COUNT'].cumsum()
df['TOTAL_DEATH'] = df['DEATH_COUNT'].cumsum()
```

```python
df['TOTAL_SURVIVED'] = df['TOTAL_CASES'] - df['TOTAL_DEATH']
```

```python
# lambda function
df.apply(lambda row: row['CASE_COUNT'], axis=1)
```

```python
df['BK_CASE_PCT'] = df.apply(lambda row: (row['BK_CASE_COUNT']/row['CASE_COUNT']) * 100
         if row['CASE_COUNT'] > 0 else 0, axis=1)
df['BK_CASE_PCT']
```

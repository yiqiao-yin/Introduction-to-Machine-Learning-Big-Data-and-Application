```python
import imageio as iio
```

```python
cat_1 = iio.imread(pth)
```

```python
cat_1.shape
```

width=159, height=318, rgb

```python
cat_1[:, :, 0].shape
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(1, 4, 1)
plt.imshow(cat_1)
plt.title('original pic')
plt.subplot(1, 4, 2)
plt.imshow(cat_1[:, :, 0], cmap='binary')
plt.title('red channel')
plt.subplot(1, 4, 3)
plt.imshow(cat_1[:, :, 1], cmap='binary')
plt.title('green channel')
plt.subplot(1, 4, 4)
plt.imshow(cat_1[:, :, 2], cmap='binary')
plt.colorbar()
plt.title('blue channel')
plt.show()
```
